```python
class Product:
    def __init__(self,name,quantity):
        self.name = name
        self.quantity = quantity
        
    def add_stock(self,qty):
        self.quantity += qty
        print(qty,"Items added.")

    def sell_product(self,qty):
        if qty <= self.quantity:
            self.quantity -= qty
            print(qty,"Items sold.")
        else:
            print("Not enough stock!")

    def show_inventory(self):
        print("Product:",self.name)
        print("Stock:",self.quantity)

# Main Program
p1 = Product("Laptop",20)
p1.show_inventory()
p1.add_stock(10)
p1.sell_product(5)
p1.show_inventory()
```
